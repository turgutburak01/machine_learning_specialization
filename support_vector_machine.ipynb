```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("tumor_dataset.csv")
data.head()
```

```python
data.drop(['id', 'Unnamed: 32'], axis=1, inplace=True)
data.head()
```

```python
M = data[data.diagnosis == 'M']
B = data[data.diagnosis == 'B']
```

```python
plt.scatter(M.radius_mean, M.texture_mean, color='red', label='kotu', alpha=0.5)
plt.scatter(B.radius_mean, B.texture_mean, color='green', label='iyi', alpha=0.5)
plt.legend()
plt.xlabel('Radius Mean')
plt.ylabel('Texture Mean')
plt.show()
```

```python
data_diagnosis = [1 if each == 'M' else 0 for each in data.diagnosis]
y = data.diagnosis.values
y
```

```python
x_data = data.drop(['diagnosis'], axis=1)
```

```python
x_data
```

```python
x = (x_data - np.min(x_data))/(np.max(x_data) - np.min(x_data))
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=1)
```

```python
from sklearn.svm import SVC
svm = SVC(random_state=1)
svm.fit(x_train, y_train)
```

```python
print("acc: ", svm.score(x_test,y_test))
```
