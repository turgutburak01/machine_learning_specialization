```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv("polynomial_linear_regression_dataset.csv", sep= ";")
df
```

```python
x = df.araba_fiyat.values.reshape(-1,1)
y = df.araba_max_hiz.values.reshape(-1,1)
```

```python
plt.scatter(x,y)
plt.xlabel("Fiyat")
plt.ylabel("Hız")
plt.show()
```

```python
lr = LinearRegression()
lr.fit(x,y)
```

```python
y_head = lr.predict(x)
```

```python
plt.scatter(x,y)
plt.xlabel("Fiyat")
plt.ylabel("Hız")
plt.plot(x, y_head, color= "red")
plt.show()
```

```python
lr.predict([[4000000]])
```

```python
from sklearn.preprocessing import PolynomialFeatures
polynomial_reg = PolynomialFeatures(degree= 2)
```

```python
x_polynomial = polynomial_reg.fit_transform(x)
x_polynomial
```

```python
linear_reg2 = LinearRegression()
linear_reg2.fit(x_polynomial,y)
```

```python
y_head2 = linear_reg2.predict(x_polynomial)
```

```python
plt.scatter(x,y)
plt.xlabel("Fiyat")
plt.ylabel("Hız")
plt.plot(x, y_head, color= "red", label="linear")
plt.plot(x, y_head2, color= "green", label="poly")
plt.legend()
plt.show()
```

```python
polynomial_reg = PolynomialFeatures(degree= 4)
x_polynomial = polynomial_reg.fit_transform(x)
```

```python
linear_reg = LinearRegression()
linear_reg.fit(x_polynomial,y)
```

```python
y_head3 = linear_reg.predict(x_polynomial)
```

```python
plt.scatter(x,y)
plt.xlabel("Fiyat")
plt.ylabel("Hız")
plt.plot(x, y_head, color= "red", label="linear")
plt.plot(x, y_head2, color= "green", label="poly")
plt.plot(x, y_head3, color= "black", label="poly")
plt.legend()
plt.show()
```
